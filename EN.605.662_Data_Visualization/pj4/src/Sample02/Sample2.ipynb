# Interactive Time Series Chart

```python
# import packages
import plotly.express as px
import pandas as pd
```

```python
# read country_vaccinations.csv 
df = pd.read_csv('country_vaccinations.csv')
```

```python
df.info()
```

8451 entries, 15 fields. 

```python
fig = px.line(df, x='date', y='total_vaccinations', color='country')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```
